# resampling (up & down)

```python
df_n = pd.read_csv('/content/mitbih_train.csv')
df_n
```

```python
from sklearn.utils import resample
df_1=df_n[df_n['0.000000000000000000e+00.88']==1]
df_2=df_n[df_n['0.000000000000000000e+00.88']==2]
df_3=df_n[df_n['0.000000000000000000e+00.88']==3]
df_4=df_n[df_n['0.000000000000000000e+00.88']==4]
df_0=(df_n[df_n['0.000000000000000000e+00.88']==0]).sample(n=2000,random_state=42)

df_1_upsample=resample(df_1,replace=True,n_samples=2000,random_state=123)
df_2_upsample=resample(df_2,replace=True,n_samples=2000,random_state=124)
df_3_upsample=resample(df_3,replace=True,n_samples=2000,random_state=125)
df_4_upsample=resample(df_4,replace=True,n_samples=2000,random_state=126)

df_n=pd.concat([df_0,df_1_upsample,df_2_upsample,df_3_upsample,df_4_upsample])
```

```python
df_n['0.000000000000000000e+00.88'].value_counts()
```

```python
new_y=df_n.iloc[:,-1]
df_n.drop('0.000000000000000000e+00.88',inplace=True,axis=1)
df_n
```

```python
train_df_n = df_n.values
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

xgb_model = XGBClassifier().fit(train_df_n, new_y)

# predict
xgb_y_predict = xgb_model.predict(test_x)

# accuracy score
xgb_score = accuracy_score(xgb_y_predict, y_test)

print('Accuracy score is:', xgb_score)
```

```python
from sklearn.metrics import classification_report
cr = classification_report(y_test, xgb_y_predict)
print(cr)
```

```python
from sklearn.metrics import confusion_matrix

cm = confusion_matrix(y_test, xgb_y_predict)
plt.figure(figsize=(8, 6))
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.xlabel("Predicted label")
plt.ylabel("True label")
plt.title("Confusion Matrix")
plt.show()
```

after upsampling the minority classes into 2000 and downsampling the majority class into 20000 the recall of class 1 and 2 increased.

# SMOTE

```python
from imblearn.over_sampling import SMOTE
# Resampling the minority class. The strategy can be changed as required.
sm = SMOTE()
# Fit the model to generate the data.
oversampled_X, oversampled_Y = sm.fit_resample(df_train.drop('0.000000000000000000e+00.88', axis=1), df_train['0.000000000000000000e+00.88'])
```

```python
oversampled_Y.value_counts()
```

```python
df_oversampled = oversampled_X.values
```

```python
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score

xgb_model = XGBClassifier().fit(df_oversampled, oversampled_Y)

# predict
xgb_y_predict = xgb_model.predict(test_x)

# accuracy score
xgb_score = accuracy_score(xgb_y_predict, y_test)

print('Accuracy score is:', xgb_score)
```

## MLP




```python
from keras.utils import to_categorical
trainY = to_categorical(new_y)
testY = to_categorical(y_test)
```

```python
from tensorflow.keras.layers import Dense, Input, Activation
from tensorflow.keras.optimizers import SGD,Adam
from tensorflow.keras.models import Model,Sequential
from sklearn.model_selection import train_test_split
```

```python
model = Sequential()
model.add(Dense(187,activation='relu' , input_shape=(187,)))
model.add(Dense(100 , activation='relu'))
model.add(Dense(5,activation='softmax'))

model.compile(optimizer='adam',
             loss='categorical_crossentropy',
             metrics=['accuracy'])

his = model.fit(train_df_n , trainY ,
                batch_size=128,
                validation_data=(test_x, testY) ,
                epochs=100 , verbose=True)
print('evaloution: ' )

loss , acc = model.evaluate(test_x , testY)
print('loss:' , loss)
print('accuracy: ' , acc)
```

```python
import numpy as np

y_pred1 = np.argmax(model.predict(test_x), axis=-1)
y_pred1
```

```python
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
# Generate a classification report
class_report = classification_report(y_test, y_pred1)
print("Classification Report:\n", class_report)
```

```python
fig , ax= plt.subplots(1,2,figsize=(10,3))
for i, met in enumerate(['accuracy', 'loss']):
    ax[i].plot(his.history[met])
    ax[i].plot(his.history['val_' + met])
    ax[i].set_title('model {}'.format(met))
    ax[i].set_xlabel('epochs')
    ax[i].set_ylabel(met)
    ax[i].legend(['train' , 'val'])
plt.show()
```

## 1D CNNs

```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, Flatten, Dense, Dropout
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report

# Assuming you have loaded your ECG dataset and preprocessed it, X_train, X_test, y_train, y_test

# Define the architecture of the 1D CNN
model = Sequential([
    Conv1D(filters=64, kernel_size=3, activation='relu', input_shape=(train_df_n.shape[1], 1)),
    MaxPooling1D(pool_size=2),
    Conv1D(filters=32, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    Conv1D(filters=16, kernel_size=3, activation='relu'),
    MaxPooling1D(pool_size=2),
    Flatten(),
    Dense(16, activation='relu'),
    Dropout(0.5),
    Dense(5, activation='softmax')  # Assuming num_classes is the number of heartbeat categories
])

# Compile the model
model.compile(loss='sparse_categorical_crossentropy',
              optimizer='adam',
              metrics=['accuracy'])

# Reshape the input data to add a channel dimension (required for Conv1D)
X_train = train_df_n.reshape(train_df_n.shape[0], train_df_n.shape[1], 1)
X_test = test_x.reshape(test_x.shape[0], test_x.shape[1], 1)

# Train the model
model.fit(X_train, new_y, epochs=10, batch_size=32, validation_split=0.1)

model.summary()
# Evaluate the model
y_pred2 = np.argmax(model.predict(X_test), axis=-1)

# Generate classification report
report = classification_report(y_test, y_pred2)

# Print the report
print(report)
```
